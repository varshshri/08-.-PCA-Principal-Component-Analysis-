# PCA - Wine Data with KMean and Heirarchial Clustering

```python
#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#from sklearn libraries import preporcessing standard scalar, decomposition PCA, Kmean
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

#inline Visualization
%matplotlib inline
```

## Data Preprocessing

```python
#X and Y split
X =  data.iloc[:,0:13].values
y = data.iloc[:,13].values
```

```python
#drop Type with new variable

data1 = data.drop(['Type'],axis=1)
data1
```

```python
#pairplot 
sns.pairplot(data.iloc[:,0:12])
```

## Normalizing Data for Clustering


```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
df_norm = norm_func(data.iloc[:,1:])
print(df_norm)
```

## Standard Scalar

```python
# this will used for kmeans
std_df = StandardScaler().fit_transform(data1)
std_df.shape
```

```python
 print(std_df)
```

## PCA for Standardized Data

```python
# applying PCA on std_df
# we are considering 95% variance in n_components to not loose any data.

pca_std = PCA(random_state=14, n_components=3)
pca_std_df= pca_std.fit_transform(std_df)
```

```python
print(pca_std_df)
```

```python
#shape
pca_std_df.shape
```

```python
# eigenvalues..
print(pca_std.singular_values_)
```

```python
# variance containing in each formed PCA

print(pca_std.explained_variance_ratio_*100)
```

```python
# Cummulative variance ratio..
# this will give an idea of, at how many no. of PCAs, the cummulative addition of
#........variance will give much information..

cum_variance = np.cumsum(pca_std.explained_variance_ratio_*100)
cum_variance
```

##### Conclusion: by applying PCA on standardized data with 95% variance it gives 3 PCA components

# Kmean Clustering with PCA


```python
#plot

plt.figure(figsize=(10, 8))
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_) #criterion based on which K-means clustering works
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method - K-means with PCA Clustering')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(df_norm)

y_kmeans
```

```python
#plot
plt.figure(figsize=(8,6))
plt.scatter(pca_std_df[:,0], pca_std_df[:,1], cmap='tab10')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

```python
pca_kmeans =pd.concat([data1.reset_index(drop = True), pd.DataFrame(df_norm)], axis= 1)
pca_kmeans.columns.values[-3: ] = ['Component 1', 'Component 2', 'Component 3']
pca_kmeans['Segment Kmeans PCA'] = y_kmeans
print(pca_kmeans)
```

```python
pca_kmeans['Segment'] = pca_kmeans['Segment Kmeans PCA'].map({0:'First', 1:'Second', 2:'Third'})
```

```python
x_axis = pca_kmeans['Component 2']
y_axis = pca_kmeans['Component 1']
plt.figure(figsize= (10,8))
sns.scatterplot(x_axis, y_axis, hue=pca_kmeans['Segment'],palette=['g', 'r','c'] )
plt.title('Clusters by PCA Componets')
plt.show()
```

```python
x_axis = pca_kmeans['Component 2']
y_axis = pca_kmeans['Component 3']
plt.figure(figsize= (10,8))
sns.scatterplot(x_axis, y_axis, hue=pca_kmeans['Segment'],palette=['g', 'r','c'] )
plt.title('Clusters by PCA Componets')
plt.show()
```

```python
x_axis = pca_kmeans['Component 1']
y_axis = pca_kmeans['Component 3']
plt.figure(figsize= (10,8))
sns.scatterplot(x_axis, y_axis, hue=pca_kmeans['Segment'],palette=['g', 'r','c'] )
plt.title('Clusters by PCA Componets')
plt.show()
```
